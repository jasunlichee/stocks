```python
from pathlib import Path
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Ideas:

Recent price change in last 1, 5, 10, 20 days
Moving average distance over last 5, 10, 20, 50 days
Volatility over last 5, 10, 20 days

1. take out 20% of the data (test)

2. Write generic curate function to curate dataset (training)

3. add the file name as a 'Ticker' column

4. merge into one dataset

5. sort by 'Date', then 'Ticker'

6. train the model


```python
LOOKBACK = 50

FEATURE_COLUMNS = [
    "return_1",
    "return_5",
    "return_10",
    "return_20",
    "ma_dist_5",
    "ma_dist_10",
    "ma_dist_20",
    "ma_dist_50",
    "volatility_5",
    "volatility_10",
    "volatility_20",
    "volume_change",
    "volume_ratio_20",
    "intraday_return",
    "range_pct",
    "close_position",
]


def add_features(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df = df.sort_values("Date").reset_index(drop=True)

    # Returns / momentum
    df["return_1"] = df["Close"].pct_change(1)
    df["return_5"] = df["Close"].pct_change(5)
    df["return_10"] = df["Close"].pct_change(10)
    df["return_20"] = df["Close"].pct_change(20)

    # Moving average distance
    ma_5 = df["Close"].rolling(5).mean()
    ma_10 = df["Close"].rolling(10).mean()
    ma_20 = df["Close"].rolling(20).mean()
    ma_50 = df["Close"].rolling(50).mean()

    df["ma_dist_5"] = (df["Close"] - ma_5) / ma_5
    df["ma_dist_10"] = (df["Close"] - ma_10) / ma_10
    df["ma_dist_20"] = (df["Close"] - ma_20) / ma_20
    df["ma_dist_50"] = (df["Close"] - ma_50) / ma_50

    # Volatility
    df["volatility_5"] = df["return_1"].rolling(5).std()
    df["volatility_10"] = df["return_1"].rolling(10).std()
    df["volatility_20"] = df["return_1"].rolling(20).std()

    # Volume features
    df["volume_change"] = df["Volume"].pct_change(1)
    volume_ma_20 = df["Volume"].rolling(20).mean()
    df["volume_ratio_20"] = df["Volume"] / volume_ma_20

    # Intraday structure
    df["intraday_return"] = (df["Close"] - df["Open"]) / df["Open"]
    df["range_pct"] = (df["High"] - df["Low"]) / df["Close"]

    day_range = df["High"] - df["Low"]
    df["close_position"] = np.where(
        day_range != 0,
        (df["Close"] - df["Low"]) / day_range,
        0.5
    )

    # Target: does tomorrow close higher than today?
    df["target"] = (df["Close"].shift(-1) > df["Close"]).astype(int)

    return df


def curate_stock_file(
    input_csv: str | Path,
    train_output_csv: str | Path,
    test_output_csv: str | Path,
    test_ratio: float = 0.20,
    lookback: int = LOOKBACK,
) -> tuple[pd.DataFrame, pd.DataFrame]:
    input_csv = Path(input_csv)
    train_output_csv = Path(train_output_csv)
    test_output_csv = Path(test_output_csv)

    df = pd.read_csv(input_csv)
    df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
    df = df.sort_values("Date").reset_index(drop=True)

    n = len(df)
    split_idx = int(n * (1 - test_ratio))

    if split_idx <= lookback:
        raise ValueError(
            f"{input_csv.name} does not have enough rows for a {lookback}-day lookback."
        )

    train_raw = df.iloc[:split_idx].copy()
    test_raw = df.iloc[split_idx:].copy()

    # Curate training set normally.
    train_curated = add_features(train_raw)

    # Drop first `lookback` rows (insufficient history)
    # and last row (no next-day target inside training set).
    train_curated = train_curated.iloc[lookback:-1].reset_index(drop=True)

    # For test: prepend last `lookback` rows from training so early test rows
    # can compute rolling features without leaking future test information backward.
    history = train_raw.iloc[-lookback:].copy()
    test_with_history = pd.concat([history, test_raw], ignore_index=True)

    test_curated = add_features(test_with_history)

    # Keep only rows that actually belong to the test period.
    first_test_date = test_raw["Date"].iloc[0]
    test_curated = test_curated[test_curated["Date"] >= first_test_date].copy()

    # Drop the last row of the test set too, because it has no t+1 target
    # inside the file.
    test_curated = test_curated.iloc[:-1].reset_index(drop=True)

    # Add ticker from filename if desired
    ticker = input_csv.stem
    train_curated["Ticker"] = ticker
    test_curated["Ticker"] = ticker

    # Reorder columns a bit
    base_columns = ["Date", "Ticker", "Open", "High", "Low", "Close", "Volume"]
    ordered_columns = base_columns + FEATURE_COLUMNS + ["target"]

    train_curated = train_curated[ordered_columns]
    test_curated = test_curated[ordered_columns]

    train_output_csv.parent.mkdir(parents=True, exist_ok=True)
    test_output_csv.parent.mkdir(parents=True, exist_ok=True)

    train_curated.to_csv(train_output_csv, index=False)
    test_curated.to_csv(test_output_csv, index=False)

    return train_curated, test_curated


def curate_directory(
    raw_dir: str | Path,
    curated_dir: str | Path,
    test_ratio: float = 0.20,
    lookback: int = LOOKBACK,
) -> None:
    raw_dir = Path(raw_dir)
    curated_dir = Path(curated_dir)

    train_dir = curated_dir / "train"
    test_dir = curated_dir / "test"

    csv_files = sorted(raw_dir.glob("*.csv"))

    for csv_file in csv_files:
        train_output_csv = train_dir / csv_file.name
        test_output_csv = test_dir / csv_file.name

        try:
            train_df, test_df = curate_stock_file(
                input_csv=csv_file,
                train_output_csv=train_output_csv,
                test_output_csv=test_output_csv,
                test_ratio=test_ratio,
                lookback=lookback,
            )
            print(
                f"Processed {csv_file.name}: "
                f"train={len(train_df)} rows, test={len(test_df)} rows"
            )
        except Exception as e:
            print(f"Skipped {csv_file.name}: {e}")


if __name__ == "__main__":
    raw_dir = "stock_market_data/sp500/raw_csv"
    curated_dir = "stock_market_data/sp500/curated_csv"

    curate_directory(raw_dir, curated_dir)
```

```python
train_dir = Path("stock_market_data/sp500/curated_csv/train")

dfs = []

for file in train_dir.glob("*.csv"):
    df = pd.read_csv(file)
    dfs.append(df)

full_train_df = pd.concat(dfs, ignore_index=True)
```

```python
full_train_df["Date"] = pd.to_datetime(full_train_df["Date"])

full_train_df = full_train_df.sort_values(
    ["Date", "Ticker"]
).reset_index(drop=True)
```

```python
FEATURES = [
    "return_1",
    "return_5",
    "return_10",
    "return_20",
    "ma_dist_5",
    "ma_dist_10",
    "ma_dist_20",
    "ma_dist_50",
    "volatility_5",
    "volatility_10",
    "volatility_20",
    "volume_change",
    "volume_ratio_20",
    "intraday_return",
    "range_pct",
    "close_position",
]
```

```python
full_train_df = full_train_df.replace([np.inf, -np.inf], np.nan)
full_train_df = full_train_df.dropna(subset=FEATURES + ["target"]).reset_index(drop=True)

X = full_train_df[FEATURES]
y = full_train_df["target"]
```

```python
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    n_jobs=-1,
    random_state=42
)

model.fit(X, y)
```

```python
test_dir = Path("stock_market_data/sp500/curated_csv/test")

dfs = []
for file in test_dir.glob("*.csv"):
    df = pd.read_csv(file)
    dfs.append(df)

full_test_df = pd.concat(dfs, ignore_index=True)

full_test_df["Date"] = pd.to_datetime(full_test_df["Date"])
full_test_df = full_test_df.sort_values(["Date", "Ticker"]).reset_index(drop=True)

full_test_df = full_test_df.replace([np.inf, -np.inf], np.nan)
full_test_df = full_test_df.dropna(subset=FEATURES + ["target"]).reset_index(drop=True)
```
